```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
# Shuffle the data
shuffled_indices = np.arange(train_data.shape[0])
np.random.shuffle(shuffled_indices)
train_data = train_data.sample(frac = 1)
train_data = train_data.drop(["Name", "PassengerId", "Ticket", "Embarked", "Cabin"], axis = 1)
test_data = test_data.drop(["Name", "PassengerId", "Ticket", "Embarked", "Cabin"], axis = 1)
```

```python
train_data.head()
```

```python
# Dealing with null values
for i in train_data.columns:
    if(train_data[i].isnull().values.any()):
        print("Data in", i, "have missing values")

age_average = train_data["Age"].mean().round(2)
train_data["Age"] = train_data["Age"].fillna(age_average)

for i in test_data.columns:
    if(test_data[i].isnull().values.any()):
        print("Data in", i, "have missing values")
test_data["Age"] = test_data["Age"].fillna(age_average)
fare_average = train_data["Fare"].mean().round(4)
test_data["Fare"] = test_data["Fare"].fillna(fare_average)
```

```python
train_data["Sex"][train_data["Sex"] == "male"] = 1
train_data["Sex"][train_data["Sex"] == "female"] = 0

test_data["Sex"][test_data["Sex"] == "male"] = 1
test_data["Sex"][test_data["Sex"] == "female"] = 0

train_data["Sex"], train_data.shape, test_data.shape
```

```python
# Create dummies variables
train_data_with_dummies = pd.get_dummies(train_data, columns = ["Pclass"], drop_first = True)
test_data_with_dummies = pd.get_dummies(test_data, columns = ["Pclass"], drop_first = True)
train_data_with_dummies.head()
```

```python
# Split train data into inputs and targets
train_inputs = train_data_with_dummies.drop(["Survived"], axis = 1)
train_targets = pd.DataFrame(columns = ["Survived"],data = train_data_with_dummies["Survived"])
```

```python
len(train_targets[train_targets["Survived"] == 1]), len(train_targets[train_targets["Survived"] == 0])
```

```python
# Balance the dataset
num_one_targets = int(np.sum(train_targets))
zero_targets = 0
indices_to_remove = []
for i in range(train_targets.shape[0]):
    if(train_targets["Survived"][i] == 0):
        zero_targets += 1
        if(zero_targets > num_one_targets):
            indices_to_remove.append(i)
# train_inputs_equal = train_inputs.drop(indices_to_remove, axis = 0)
# train_targets_equal = train_targets.drop(indices_to_remove, axis = 0)
train_inputs_equal = train_inputs
train_targets_equal = train_targets
len(train_targets_equal[train_targets_equal["Survived"] == 1]), len(train_targets_equal[train_targets_equal["Survived"] == 0])
```

```python
# Train test split
from sklearn.model_selection import train_test_split
train_inputs, val_inputs, train_targets, val_targets = train_test_split(train_inputs_equal, train_targets_equal, test_size = 0.2)
```

```python
# Standardize the inputs
from sklearn import preprocessing
scaled_train_inputs = preprocessing.scale(train_inputs)
scaled_val_inputs = preprocessing.scale(val_inputs)
scaled_test = preprocessing.scale(test_data_with_dummies)
```

```python
# Save data
np.savez("Titanic data train", inputs = scaled_train_inputs, targets = train_targets)
np.savez("Titanic data test", inputs = scaled_test)
np.savez("Titanic data validation", inputs = scaled_val_inputs, targets = val_targets)
```

```python
# Load data
npz = np.load("Titanic data train.npz")
train_inputs = npz["inputs"]
train_targets = npz["targets"]

npz = np.load("Titanic data validation.npz")
val_inputs = npz["inputs"]
val_targets = npz["targets"]
```

```python
input_size = train_inputs.shape[1]
output_size = 2
hidden_layer_size = 50

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_shape = (input_size,), activation = "relu"),
    tf.keras.layers.Dense(hidden_layer_size, activation = "relu"),
    tf.keras.layers.Dense(hidden_layer_size, activation = "relu"),
    tf.keras.layers.Dense(hidden_layer_size, activation = "relu"),
    tf.keras.layers.Dense(output_size, activation = "softmax")
])

model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["acc"])

# Training
batch_size = 100
max_epochs = 100
early_stopping = tf.keras.callbacks.EarlyStopping(patience = 1)
model.fit(train_inputs,
          train_targets,
          batch_size = batch_size,
          validation_split = 0.15,
          epochs = max_epochs,
          callbacks = [early_stopping],
          verbose = 2)
```

```python
data_test = np.load("Titanic data test.npz")["inputs"]
test_targets = model.predict([data_test]).round(2)
print(test_targets)
predictions = []
for i in test_targets:
    if(i[1] > 0.5):
        predictions.append(1)
    else:
        predictions.append(0)
predictions = np.array(predictions)
```

```python
test_loss, test_accuracy = model.evaluate(data_test, predictions)
```

```python
test_data = pd.read_csv("test.csv")
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
